## Data Cleaning

```python
music.columns = music.columns.str.strip()
music["songname_artist"] = music["name"] + '-' + music["artist"]
music.set_index('songname_artist', inplace=True)
music = music.drop(['name', 'artist',  'liveness', 'type', 'mode', 'duration_ms', 'tempo', 'id', 'html', 'time_signature'], axis =1)
df_5000 = music.iloc[:,:13]
df_5000
```

## Data Transformation

```python
#standardize data    'danceability*' 'energy*' 'key*' 'loudness*' 'speechiness*' 'acousticness*' 'instrumentalness' 'valence' 
columns = ['acousticness', 'instrumentalness']
df_5000sample = df_5000.filter(columns) #choose collumns here 
df_5000norm = MinMaxScaler().fit_transform(df_5000sample)
pd.DataFrame(df_5000norm).head()
```

## Calculate KMeans and filter

```python
#calculate cluster centers
kmeans = KMeans(n_clusters=20).fit(df_5000norm)  #define clusters here

kmeans.cluster_centers_
centroids = pd.DataFrame(kmeans.cluster_centers_)
centroids.columns = columns
centroids
```

```python
# Make a dataframe of the normalised data to plot on graph
plot_data = pd.DataFrame(df_5000norm)
plot_data.columns = columns

# Add a column to say which cluster a penguin belongs to
plot_data['cluster'] = kmeans.labels_

# read the species as a column so it can be seen on the chart
#plot_data = pd.concat([plot_data, penguins.species], axis=1)

plot_data.head()
```

## Inerta Calculation

```python
kmeans.inertia_
```

```python
max_k = 30
inertia = []

for k in range(1, max_k): 
    inertia.append(KMeans(n_clusters=k).fit(df_5000norm).inertia_)

fig, ax = plt.subplots(figsize=(20, 10))
plt.title(f'Penguins inertia evolution from 1 cluster to {max_k} cluster')
sns.lineplot(x=range(1, max_k), y=inertia)
sns.scatterplot(x=range(1, max_k), y=inertia)
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
silhouette_score(df_5000norm, kmeans.labels_)
```

```python
max_k = 70
sil_score = []

for k in range(2, max_k): 
    labels = KMeans(n_clusters=k).fit(df_5000norm).labels_
    sil_score.append(silhouette_score(df_5000norm, labels))

plt.title(f'Silhouette score evolution from 2 cluster to {max_k} clusters')
sns.lineplot(x=range(2, max_k), y=sil_score)
sns.scatterplot(x=range(2, max_k), y=sil_score)
plt.show()
```

```python
max_k = 12
data = df_5000norm

for k in range(1, max_k): 
    kmean_model = KMeans(n_clusters=k).fit(data)
    labels = kmean_model.labels_
    
    if 1 < k < data.shape[0]: 
        
        fig, (axis1, axis2) = plt.subplots(1, 2)
        fig.set_size_inches(20, 8)
        
        sil_avg = silhouette_score(data, labels)
        print(f"* For k = {k} the average to silhouette is: {round(sil_avg,4)}")
        sample_sil_val = silhouette_samples(data, labels)
        
        y_lower = 10
        
        for i in range(k): 
            ith_cluster_sv = sample_sil_val[labels == i]
            print(f"\t- For cluster = {i} the silhouette value is: {round(np.mean(ith_cluster_sv),2)}")
            
            ith_cluster_sv.sort()
            
            # Calculate where to put the silhuette on the y axes
            ith_cluster_size = ith_cluster_sv.shape[0]
            y_upper = y_lower + ith_cluster_size
 
            
            # Paint the cluster
            axis1.fill_betweenx(np.arange(y_lower, y_upper),
                               0, ith_cluster_sv, 
                                alpha = 0.7)
            
            
            # Label the cluster
            axis1.text(-0.05, y_lower + 0.5 * ith_cluster_size, str(i))
            
            # Calculate the next y lower value for the next cluster of the graph
            y_lower = y_upper + 10 #dejamos vacías 10 posiciones sin muestra
            
        axis1.set_title("Silhouette score for k = %s"%str(k))
        axis1.set_xlabel("S(i)")
        axis1.set_ylabel("Cluster ID")

    if 1 < k: 
        centr_df_temp = pd.DataFrame(kmean_model.cluster_centers_)
        centr_df_temp.columns = columns
        centr_df_temp['cluster_name'] = centr_df_temp.index

        plot_data = pd.DataFrame(data)
        plot_data.columns = columns
        plot_data['cluster'] = kmean_model.predict(data)

        plt.title('Songs by acousticness and instrumentalness')
        sns.scatterplot(data=plot_data, x= 'acousticness', y='instrumentalness', hue='cluster', palette='Set2')
        sns.scatterplot(data=centr_df_temp, x= 'acousticness', y='instrumentalness', color='red', s=150)
        for i in range(centr_df_temp.shape[0]):
            plt.text(
                x=centr_df_temp.acousticness[i], 
                y=centr_df_temp.instrumentalness[i],
                s=centr_df_temp.cluster_name[i], 
                size=20
            )
        plt.show()
```
